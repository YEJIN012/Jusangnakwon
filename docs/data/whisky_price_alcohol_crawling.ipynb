```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from urllib.parse import quote
import urllib.request
import cloudscraper
from currency_converter import CurrencyConverter
from currency_converter import SINGLE_DAY_ECB_URL

drink_csv = pd.read_csv('/content/drive/MyDrive/drink_data/whisky.csv',encoding='CP949')

c = CurrencyConverter(SINGLE_DAY_ECB_URL)
prices = []
alcohols = []
cnt=0
for index, row in drink_csv.iterrows():

    scraper = cloudscraper.create_scraper(delay=10,   browser={'custom': 'ScraperBot/1.0',})
    url = row['link']
    req = scraper.get(url)
    soup = BeautifulSoup(req.text,'lxml')
    price_ptag = soup.select(".block-price > p")
    if len(price_ptag)!=2:
        price = ""
    else:
        price_ptag = soup.select(".block-price > p")[1]
        if price_ptag is not None:
                price = price_ptag.get_text()[2:]
                try:
                    price = c.convert(price, 'EUR', 'KRW')
                except ValueError:
                    price = ""
        else:
                price = ""

    alcohol_tag = soup.find('dd', text=lambda x: x and '% Vol.' in x)

    if alcohol_tag is not None:
            alcohol = alcohol_tag.get_text()[:-7]
    else:
            alcohol = ""

    

    prices.append(price)
    alcohols.append(alcohol)
    cnt=cnt+1
    print(cnt)

drink_csv['price'] = prices
drink_csv['alcohol'] = alcohols
```

```python
drink_csv.to_csv("whisky_res.csv", index = False)
```
